```python
import seaborn as sns
```

```python
df.set_index('País',inplace=True)
```

```python
df
```

```python
top_10 = df.sort_values(by='Total',ascending=False).head(10)
```

```python
top_10
```

```python
import matplotlib.pyplot as plt
```

```python
# Tema + Paleta = Seaborn
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h')
ax.set(title='Os dez paises com mais imigrantes para o canadá',xlabel='total de imigrantes',ylabel='')
plt.show()
```

```python
def gera_grafico_paleta(pallete):
  fig,ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h',palette=pallete)
  ax.set_title('os des paises com mais imigrantes para o canadá',loc='left',fontsize=16)
  ax.set_xlabel('Total de imigrantes',fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
gera_grafico_paleta('Blues')
```

```python
gera_grafico_paleta('tab10')
```

```python
gera_grafico_paleta("Spectral")
```
